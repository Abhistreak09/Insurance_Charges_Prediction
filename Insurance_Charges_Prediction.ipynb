## Data Pre-processing (Splitting)


```python
x = data.drop(['region','charges'], axis = 1)
y = data.charges
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=2)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
x_train
```

## Model Building & Evaluation

```python
from sklearn.linear_model import LinearRegression 
reg = LinearRegression() 

reg.fit(x_train,y_train)
```

```python
# Making Prediction on train data

y_p = reg.predict(x_train)
```

```python
print('R2 score: ', r2_score(y_train, y_p))
```

```python
# Making Prediction on test data

y_pred = reg.predict(x_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R2 score: ', r2_score(y_test, y_pred))
```

##### Result of my model is able to expain 84% for the total data set.
